```python
import numpy as np

import statsmodels.formula.api as smf
import patsy
```

```python
for i in list(df.columns):
    display(df[i].value_counts().to_frame())
```

```python
reg=smf.ols('''
            np.log(renda) ~ tempo_emprego + C(sexo, Treatment(0)) + C(posse_de_veiculo, Treatment(0))
            + C(posse_de_imovel, Treatment (0)) + C(qtd_filhos, Treatment(0)) + C(tipo_renda, Treatment(0))
             + C(educacao, Treatment(0)) + C(estado_civil, Treatment(0)) + C(tipo_residencia, Treatment(0))
             + idade + C(qt_pessoas_residencia, Treatment(2))
            ''',df).fit()
reg.summary()
```

```python
reg=smf.ols('''
            np.log(renda) ~ tempo_emprego + C(sexo, Treatment(0)) + C(posse_de_veiculo, Treatment(0))
            + C(posse_de_imovel, Treatment (0)) + C(tipo_renda, Treatment(0))
            + C(estado_civil, Treatment(0)) + idade + C(qt_pessoas_residencia, Treatment(2))
            ''',df).fit()
reg.summary()
```

De acordo com os indicadores o modelo não parece melhor, observa-se um queda na $R^2$ ajustado de 0,357 para 0,353 e também nota-se um aumento no AIC de 2,719e+04 para 2,724e+04
