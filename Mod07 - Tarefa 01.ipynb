#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável. 
    Dica: 
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis, 
        3. conte o número de categorias dessa variável 
        4. acumule essa informação de 3. na lista que você criou em 1. 
        5. No final, essa lista pode ser atribuída à nossa variável.
3. Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.

````python
#1)
df_tipos=pd.DataFrame(df.dtypes, columns=["tipos"])
df_tipos
````

````python
#2)
qtd_categorias=[]

for col in df.columns:    
    num_categorias = df[col].nunique()    
    qtd_categorias.append(num_categorias)    
     
df_tipos['qtd_categorias']=qtd_categorias
df_tipos
````

````python
#3)
df_tipos= pd.get_dummies(df_tipos, columns=['tipos'])
df_tipos
````

````python
df_tipos.tipos_int64=df_tipos.tipos_int64.astype(int)
df_tipos.tipos_float64=df_tipos.tipos_float64.astype(int)
df_tipos.tipos_object=df_tipos.tipos_object.astype(int)
df_tipos.tipos_bool=df_tipos.tipos_bool.astype(int)
df_tipos
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
df_dummies= pd.get_dummies(df, columns=['tipo_renda','educacao','estado_civil','tipo_residencia'],drop_first=True)
df_dummies.columns
````

````python
df_dummies.columns=['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'qtd_filhos', 'idade',
       'tempo_emprego', 'possui_celular', 'possui_fone_comercial',
       'possui_fone', 'possui_email', 'qt_pessoas_residencia', 'mau',
       'tipo_renda_Pensioner', 'tipo_renda_State_servant',
       'tipo_renda_Student', 'tipo_renda_Working', 'educacao_Higher_education',
       'educacao_Incomplete_higher', 'educacao_Lower_secondary',
       'educacao_Secondary_secondary_special', 'estado_civil_Married',
       'estado_civil_Separated', 'estado_civil_Single_not_married',
       'estado_civil_Widow', 'tipo_residencia_House_apartment',
       'tipo_residencia_Municipal_apartment',
       'tipo_residencia_Office_apartment', 'tipo_residencia_Rented_apartment',
       'tipo_residencia_With_parents']
df_dummies.columns
````

````python

df_dummies.sexo = df_dummies.sexo.map({'F': 1, 'M':0})
df_dummies.posse_de_veiculo = df_dummies.posse_de_veiculo.map({'Y': 1, 'N':0})
df_dummies.posse_de_imovel = df_dummies.posse_de_imovel.map({'Y': 1, 'N':0})

df_dummies = df_dummies.apply(lambda x: x.astype(int) if x.dtype == 'bool' else x)
df_dummies.dtypes
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
freq = pd.crosstab(df_dummies['possui_email'], df_dummies['posse_de_veiculo'])
freq
````

posse_de_veiculo é mais poderosa pois possue maior quantidade de valores
