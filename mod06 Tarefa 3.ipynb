```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook

df = pd.read_csv('b1abb33c4f.csv')
df
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[(df['APGAR5'] >= 8) & (df['APGAR5'] <= 10),
           'apgar5_cat'] = 'normal'
df.loc[(df['APGAR5'] >= 6) & (df['APGAR5'] <= 7),
           'apgar5_cat'] = 'asfixia leve'
df.loc[(df['APGAR5'] >= 4) & (df['APGAR5'] <= 5),
           'apgar5_cat'] = 'asfixia moderada'
df.loc[(df['APGAR5'] >= 0) & (df['APGAR5'] <= 3),
           'apgar5_cat'] = 'asfixia grave'
df
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.catplot(x='SEXO', y='APGAR5'
            , hue = 'GESTACAO'
            , kind='bar'
            , data=df
            , errorbar=None
           )
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
freq = pd.crosstab(df['GESTACAO'], df['APGAR5'])
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_pctlin = freq.div(freq.sum(axis = 1), axis = 0)
freq_pctlin
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
freq_pctlin.plot.bar()
```

O conjunto de dados corrobora com a afirmação dos médicos, quanto menor a gestação aumenta a frequencia de APGAR5 de valor baixo
