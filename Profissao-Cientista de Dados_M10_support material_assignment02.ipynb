````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

%matplotlib inline

warnings.filterwarnings("ignore")

df=pd.read_csv('Profissao- Cientista de Dados_M10_support material.csv')
df
````

````python
df1=df[['posse_de_imovel','posse_de_veiculo','renda']]
df1.corr()
````

````python
cmap = sns.diverging_palette(h_neg=125, h_pos=350, as_cmap=True, sep = 60, center = 'light')

sns.clustermap(df1.corr(), figsize=(6, 6), center = 0, cmap=cmap)
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=df1, dodge=True, ci = 90)
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df1, dodge=True, ci = 90)
````

Ambas variáveis possuem correlações baixas, sendo a correlação de posse de veículo maior. Os gráficos de perfil de média demonstram que quem possui um veiculo ou imovel tem possui a renda maior.
